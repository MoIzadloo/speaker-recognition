```python
import numpy as np
import tensorflow as tf
import pandas as pd
import ast
```

```python
sequence_length = 50
n_mfcc = 20
```

```python
def extract_mfccs(path):
    df = pd.read_csv(path)
    x = []
    ids = df["speaker_id"]
    mfccs = df["mfcc"]
    x = []
    y = np.array([])
    for i in range(len(mfccs)):
        mfcc = mfccs.apply(ast.literal_eval).to_numpy()[i]
        mfcc /= np.max(mfcc)
        for j in range(0, len(mfcc), sequence_length):
            sampleX = np.array(mfcc[j: j + sequence_length])
            try:
                sampleX.reshape(sequence_length, n_mfcc)
                x.append(sampleX)
                y = np.append(y, ids[i])
            except:
                continue

    x = tf.reshape(x, shape=[-1, sequence_length, n_mfcc])
    return [x, y]

x, y = extract_mfccs("samples_dataset.csv")
x.shape
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, return_sequences=True, input_shape=(sequence_length, n_mfcc)),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x, y, epochs=10, batch_size=32)
```

```python
x, y = extract_mfccs("tests_dataset.csv")
model.evaluate(x, y)
```

```python
output = model.predict(x)
list(zip(output, y))
```
